```python
from popper.utils import ExperimentalDataLoader
from popper.benchmark import gene_perturb_hypothesis
from popper.agent import SequentialFalsificationTest

data_path = 'YOUR PATH TO THE DATA'
data_loader = ExperimentalDataLoader(data_path, table_dict_selection = 'all_bio')
```

```python
for i,j in data_loader.table_dict.items():
    display(j.head(1))
```

```python
bm = gene_perturb_hypothesis(num_of_samples = 1, permuted=False, dataset = 'IL2', path = data_path)
example = bm.get_example(0)
```

```python
example
```

```python
agent = SequentialFalsificationTest(llm = "claude-3-5-sonnet-20241022")

agent.configure(data = data_loader, alpha = 0.1, aggregate_test = 'E-value', 
            max_num_of_tests = 5, max_retry = 5, time_limit = 2, 
            relevance_checker = True, use_react_agent=True)

log, last_message, parsed_result = agent.go(example['prompt'])
```
